```python
# Project 3: Exploratory Data Analysis (EDA) - Steam Games Dataset
# Name: Aditya Shibu
# Roll Number: 2401201047
# Class & Section: SEC B
# Course: BCA AI&DS
# Subject: Probabilistic Modelling and Reasoning (ENSP205)


# ========================
# 1️⃣ Import Libraries
# ========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Plot style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12,6)
```

```python
# ========================
# 2️⃣ Load Dataset
# ========================

# ========================
# Source = https://www.kaggle.com/datasets/nikdavis/steam-store-games
# ========================
file_path = r"C:\Users\Agney Shibu\Downloads\archive\steam.csv"
df = pd.read_csv(file_path)

# Inspect the dataset
display(df.head())
display(df.tail())
df.info()
df.describe(include='all')
```

```python
# ========================
# 3️⃣ Data Cleaning
# ========================

# Missing values
print("Missing values per column:\n", df.isnull().sum())

# Fill or drop missing values
df = df.dropna(subset=['price', 'genres'])  # critical columns
df['publisher'] = df['publisher'].fillna('Unknown')

# Remove duplicates
df = df.drop_duplicates()

# Detect outliers in numeric columns
numeric_cols = ['price','positive_ratings','negative_ratings','average_playtime','median_playtime']
for col in numeric_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Outliers in {col}')
    plt.show()
```

```python
# ========================
# 4️⃣ Univariate Analysis
# ========================

# Numeric columns
for col in numeric_cols:
    print(f"--- {col.upper()} ---")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    trimmed_mean = df[col].sort_values()[int(0.05*len(df)):int(0.95*len(df))].mean()
    print("Trimmed Mean:", trimmed_mean)
    print("Range:", df[col].max() - df[col].min())
    print("Variance:", df[col].var())
    print("Standard Deviation:", df[col].std())
    print("\n")
    
    # Distribution plots
    plt.figure()
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()

# Categorical columns
cat_cols = ['genres','developer','publisher']
for col in cat_cols:
    print(f"\nTop 5 in {col}:")
    print(df[col].value_counts().head())
    plt.figure()
    sns.countplot(y=df[col], order=df[col].value_counts().iloc[:10].index)
    plt.title(f"Top 10 {col.title()}")
    plt.show()
```

```python
# ========================
# 5️⃣ Bivariate / Multivariate Analysis
# ========================

# ensure required derived cols exist
if 'owners_median' not in df.columns:
    df['owners_min'] = df['owners'].apply(lambda x: int(x.split('-')[0]))
    df['owners_max'] = df['owners'].apply(lambda x: int(x.split('-')[1]))
    df['owners_median'] = (df['owners_min'] + df['owners_max']) / 2

# Create log-transformed numeric columns for plotting (safe for zeros)
df['log_price'] = np.log1p(df['price'])
df['log_positive_ratings'] = np.log1p(df['positive_ratings'])
df['log_negative_ratings'] = np.log1p(df['negative_ratings'])
df['log_average_playtime'] = np.log1p(df['average_playtime'])
df['log_median_playtime'] = np.log1p(df['median_playtime'])

# Simplify genres: take the primary genre (first token before ';') to reduce cardinality
df['primary_genre'] = df['genres'].apply(lambda s: str(s).split(';')[0].strip() if pd.notnull(s) else 'Unknown')

# Select top genres for clearer categorical plots
top_genres = df['primary_genre'].value_counts().nlargest(10).index.tolist()
```

```python
# 1) Correlation heatmap (cleaned, with mask and formatting)
numeric_cols_for_corr = ['price','positive_ratings','negative_ratings','average_playtime','median_playtime','owners_median']
corr = df[numeric_cols_for_corr].corr()

# plot
plt.figure(figsize=(8,6))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5, vmin=-1, vmax=1)
plt.title("Correlation Heatmap (Numeric Features)")
plt.tight_layout()
plt.show()
```

```python
# 2) Scatter: Price vs Positive Ratings (log-log) — sampled to avoid overplotting
sample_n = min(5000, len(df))
df_sample = df.sample(n=sample_n, random_state=42)

plt.figure(figsize=(8,6))
sc = plt.scatter(np.log1p(df_sample['price']), np.log1p(df_sample['positive_ratings']),
                 c=df_sample['log_average_playtime'], cmap='viridis', alpha=0.6, s=20)
plt.colorbar(sc, label='Log(Average Playtime + 1)')
plt.xlabel('Log(Price + 1)')
plt.ylabel('Log(Positive Ratings + 1)')
plt.title('Price vs Positive Ratings (log-log, sampled)')
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
# 3) Positive vs Negative Ratings (log) — shows opinion polarity, colored by average playtime
plt.figure(figsize=(8,6))
df_sample2 = df.sample(n=5000, random_state=1)
sns.scatterplot(x='log_positive_ratings', y='log_negative_ratings',
                hue='log_average_playtime', data=df_sample2, palette='magma',
                edgecolor=None, alpha=0.6)
plt.xlabel('Log(Positive Ratings + 1)')
plt.ylabel('Log(Negative Ratings + 1)')
plt.title('Positive vs Negative Ratings (log) — color = log avg playtime')
plt.legend(title='Log Avg Playtime', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 4) Boxplot: Price distribution by Top-10 primary genres (use log-price to handle outliers)
plt.figure(figsize=(12,6))
sns.boxplot(x='primary_genre', y='log_price', data=df[df['primary_genre'].isin(top_genres)],
            order=top_genres)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Log(Price + 1)')
plt.title('Price Distribution (log) by Top 10 Primary Genres')
plt.tight_layout()
plt.show()
```

```python
# 5) Average Playtime by Top Genres (log), barplot of median log playtime
genre_stats = (df[df['primary_genre'].isin(top_genres)]
               .groupby('primary_genre')['log_average_playtime']
               .median()
               .sort_values(ascending=False))

plt.figure(figsize=(10,5))
sns.barplot(x=genre_stats.index, y=genre_stats.values)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Median Log(Average Playtime + 1)')
plt.title('Median Log Average Playtime by Top 10 Genres')
plt.tight_layout()
plt.show()
```

```python
# 6) Ratings vs Price by Top Publishers (boxplot of log positive ratings vs publisher)
top_publishers = df['publisher'].value_counts().nlargest(8).index.tolist()
plt.figure(figsize=(12,6))
sns.boxplot(x='publisher', y='log_positive_ratings', data=df[df['publisher'].isin(top_publishers)],
            order=top_publishers)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Log(Positive Ratings + 1)')
plt.title('Positive Ratings Distribution (log) by Top Publishers')
plt.tight_layout()
plt.show()
```

```python
# 7) Handy tables: top outliers for quick reporting
print("Top 10 by positive_ratings:")
display(df.sort_values('positive_ratings', ascending=False)[['name','positive_ratings','price','primary_genre']].head(10))

print("\nTop 10 by average_playtime:")
display(df.sort_values('average_playtime', ascending=False)[['name','average_playtime','price','primary_genre']].head(10))

print("\nMost expensive games:")
display(df.sort_values('price', ascending=False)[['name','price','primary_genre']].head(10))
```
